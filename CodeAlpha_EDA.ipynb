```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load dataset
data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
# Step 2: Ask meaningful questions
print("Questions to Explore:")
print("1. What is the structure of the dataset?")
print("2. Are there missing values or duplicates?")
print("3. What are the key patterns in age, fare, and class?")
print("4. How are survival rates distributed across genders and classes?\n")
```

```python
# Step 3: Explore data structure
print("Dataset Overview:")
print(data.info())
print("\n Summary Statistics:")
print(data.describe())
print("\n Data Shape:", data.shape)
```

```python
# Step 4: Check for missing or duplicate data
print("\n Missing Values:")
print(data.isnull().sum())
print("\n Duplicates:", data.duplicated().sum())
```

```python
# Step 5: Identify trends and patterns
plt.figure(figsize=(7,5))
sns.countplot(x='Survived', data=data, palette='Set2')
plt.title("Survival Count")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(data['Age'].dropna(), bins=20, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Pclass', y='Fare', data=data, palette='coolwarm')
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
# Step 6: Hypothesis testing & correlation
plt.figure(figsize=(6,5))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Step 7: Detect potential issues
print("\n Potential Data Issues:")
if data.isnull().sum().any():
    print("- Missing values found. Consider filling or dropping them.")
if data.duplicated().sum() > 0:
    print("- Duplicate records found.")
if (data['Fare'] > data['Fare'].mean() + 3*data['Fare'].std()).any():
    print("- Possible outliers detected in 'Fare' column.")
else:
    print("- No major data issues detected.")
```
